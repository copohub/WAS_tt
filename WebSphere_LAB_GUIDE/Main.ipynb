# Archivo y Nomenclatura de Rutas

WebSphere Application Server es soportado en Linux, Windows, Unix como IBM AIX, HP AUX, Sun Solaris. Este laboratorio trabajara en sistemas Linux. WebSphere Application Server para Windows incluye .bat versiones de los scripts.

Para el proposito de este laboratorio, nombre de archivos y rutas de archivos seran referenciados al estilo UNIX. Si hay alguna instrucción para realizar en Windows se especificara la ejecución de dicha tarea.
